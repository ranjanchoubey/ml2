## **Solution 1.a.  :**
<br>
<div align="center",style="margin-top: 10px;">
if $f$ is a convex function and $x_1, x_2, \ldots, x_n$ are points in its domain with non-negative weights $w_1, w_2, \ldots, w_n$ such that $\sum_{i=1}^{n} w_i = 1$, then
from  Jensens Inequality

&nbsp;

$$\therefore \boxed {
f\left(\sum_{i=1}^{n} w_i x_i\right) \leq \sum_{i=1}^{n} w_i f(x_i)}
$$

&nbsp;

 put $w_i = \frac{a_i}{\sum_{j=1}^{n} a_j}$ and $x_i = \frac{b_i}{a_i}$ into above Jensen's inequality

$$\Rightarrow
f\left(\sum_{i=1}^{n} w_i x_i\right) = f\left(\sum_{i=1}^{n} \frac{a_i}{\sum_{j=1}^{n} a_j} \cdot \frac{b_i}{a_i}\right) = f\left(\frac{\sum_{i=1}^{n} b_i}{\sum_{i=1}^{n} a_i}\right)
 \leq \sum_{i=1}^{n} \frac{a_i}{\sum_{j=1}^{n} a_j} f\left(\frac{b_i}{a_i}\right).
$$

&nbsp;

Now, Multiplying by $\sum_{i=1}^{n} a_i$ on both side
&nbsp;

$$\therefore \boxed{
\left(\sum_{i=1}^{n} a_i\right) f\left(\frac{\sum_{i=1}^{n} b_i}{\sum_{i=1}^{n} a_i}\right) \leq \sum_{i=1}^{n} a_i f\left(\frac{b_i}{a_i}\right)}
 proved !
$$

Let  $A = \int a(x) u(x) \, dx$ , $B = \int b(x) u(x) \, dx$ writing Jensen's inequality for a convex function $f$:

$$\boxed{
f\left(\frac{\int_\mathcal{X} g(x) \, d\mu(x)}{\int_\mathcal{X} h(x) \, d\mu(x)}\right) \leq \frac{\int_\mathcal{X} h(x) f\left(\frac{g(x)}{h(x)}\right) \, d\mu(x)}{\int_\mathcal{X} h(x) \, d\mu(x)}}
$$


where, $\mu$  =  measure, and $g(x)$ and $h(x)$ are non-negative functions.


&nbsp;

Let $h(x) = a(x) u(x)$ and $g(x) = b(x) u(x)$ ;
where measure is induced by the weight function $u(x)$.
&nbsp;


$$\Rightarrow
f\left(\frac{\int_\mathcal{X} b(x) u(x) \, dx}{\int_\mathcal{X} a(x) u(x) \, dx}\right) \leq \frac{\int_\mathcal{X} a(x) f\left(\frac{b(x)}{a(x)}\right) u(x) \, dx}{\int_\mathcal{X} a(x) u(x) \, dx}.
$$


&nbsp;

Multiplying both sides by $\int_\mathcal{X} a(x) u(x) \, dx$ ,we get

&nbsp;


$$\therefore \boxed{
\int a(x) u(x) \, dx \, f\left(\frac{\int b(x) u(x) \, dx}{\int a(x) u(x) \, dx}\right) \leq \int a(x) f\left(\frac{b(x)}{a(x)}\right) u(x) \, dx}
Proved !

$\because$ $f$ is convex and differentiable, so using Convexity and Lipschitz Continuity:

$$

Now, putting into the update rule: $x_{k+1} - x_k = -\frac{1}{L} \nabla f(x_k)$:

$$
\boxed{
\|x_{k+1} - x_k\| = \left\|-\frac{1}{L} \nabla f(x_k)\right\| = \frac{1}{L} \|\nabla f(x_k)\|
}
$$

&nbsp;

Putting this into the inequality:

$$
\Rightarrow
f(x_{k+1}) \leq f(x_k) - \frac{1}{L} \|\nabla f(x_k)\|^2 + \frac{L}{2} \left(\frac{1}{L} \|\nabla f(x_k)\|\right)^2
$$

$$
\Rightarrow
f(x_{k+1}) \leq f(x_k) - \frac{1}{2L} \|\nabla f(x_k)\|^2
$$

&nbsp;

Now, bounding the difference $f(x_k) - f(x^*)$ by summing the inequality from $k = 0$ to $K-1$:

$$
\sum_{k=0}^{K-1} (f(x_k) - f(x_{k+1})) \geq \frac{1}{2L} \sum_{k=0}^{K-1} \|\nabla f(x_k)\|^2.
$$

&nbsp;

This is a telescopic sum, which finally gives:

$$
f(x_0) - f(x_K) \geq \frac{1}{2L} \sum_{k=0}^{K-1} \|\nabla f(x_k)\|^2.
$$

&nbsp;

$\because f(x_K) \geq f(x^*)$

$$
\therefore
f(x_0) - f(x^*) \geq \frac{1}{2L} \sum_{k=0}^{K-1} \|\nabla f(x_k)\|^2.
$$
&nbsp;

$$
\Rightarrow
\boxed{
f(x_k) - f(x^*) \leq \frac{5L}{2} \sum_{i=0}^{k-1} \|x_i - x^*\|^{-2}
}
$$

where, the factor $\frac{5L}{2}$ comes from bounding the error terms using the gradient and convexity properties.

$$\boxed{
\therefore \; \exp(f_y) > 0}
$$

&nbsp;

Also, ${\sum_{j=1}^{k} \exp(f_j) > \exp(f_y) > 0}$


$$\therefore \boxed{
0 < \frac{\exp(f_y)}{\sum_{j=1}^{k} \exp(f_j)} < 1}
$$

&nbsp;

Taking the minus log both side, we get,

$$
-\log\left(\frac{\exp(f_y)}{\sum_{j=1}^{k} \exp(f_j)}\right) > 0.
$$

$
\therefore \boxed{ \; 0 < \ell^{CE}(f, y) < \infty } Proved!
$

**(b)** Show that  $\ell^{CE}(\lambda e_y, y) \to 0$ as $\lambda \to \infty$

&nbsp;

**Proof:**

<div align="center" style="padding: 10px;">

Let $f = \lambda e_y$, where $f_y = \lambda$ and $f_j = 0$ for $j \neq y$.

&nbsp;

The cross-entropy loss

$$\therefore \boxed{
\ell^{CE}(\lambda e_y, y) = -\log\left(\frac{\exp(\lambda)}{\exp(\lambda) + \sum_{j \neq y} \exp(0)}\right)}
$$

$$⇒
\ell^{CE}(\lambda e_y, y) = -\log\left(\frac{\exp(\lambda)}{\exp(\lambda) + (k-1)}\right).
$$

&nbsp;

As $\lambda \to \infty$, $\exp(\lambda) \gg (k-1)$:

$$\therefore\boxed{
\frac{\exp(\lambda)}{\exp(\lambda) + (k-1)} \to 1}
$$

&nbsp;

Now ,Take the log both side , we get


$$
\ell^{CE}(\lambda e_y, y) \to -\log(1) = 0
$$

&nbsp;

$$
\therefore \; \ell^{CE}(\lambda e_y, y) \to 0 \; \text{as} \; \lambda \to \infty
$$

## **Solution 6:**

Two forms of (non-Nesterov) momentum :

<div align="center" style="padding: 10px;">

$\textbf{Form I:}$ $$\boxed{
\theta^{k+1} = \theta^k - \alpha g^k + \beta (\theta^k - \theta^{k-1})}
$$

&nbsp;

$\textbf{Form II:}$ $$\boxed{
\theta^{k+1} = \theta^k - \alpha v^{k+1}}
$$

&nbsp;


$\textbf{Proof (by Induction):}$

Let us assume that the  sequence $\{\theta^k\}$ generated by Form I and Form II is identical.

**Base Case ($k = 0$):**

For Form I:
$$
\theta^0 = \theta^0
$$

For Form II:
$$
v^0 = 0 \quad \text{and} \quad \theta^0 = \theta^0
$$

$\therefore$ Both forms start with the same initial $\theta^0$.

&nbsp;


**Inductive Step: Assume equivalence holds for some $k = m$.**

We assume:
$$
\boxed{\theta^m = \theta^m} \quad \text{(from both forms)}
$$

&nbsp;

**Now Showing equivalence for $k = m + 1$.**

For Form I:
$$\boxed{
\theta^{m+1} = \theta^m - \alpha g^m + \beta (\theta^m - \theta^{m-1})}
$$

&nbsp;

For Form II:
$$
v^{m+1} = g^m + \beta v^m
$$
$$\boxed{
\theta^{m+1} = \theta^m - \alpha v^{m+1}}
$$
&nbsp;


put $v^{m+1}$ into $\theta^{m+1}$ in Form II:
&nbsp;

$$
\theta^{m+1} = \theta^m - \alpha (g^m + \beta v^m)
$$
$$
\therefore \theta^{m+1} = \theta^m - \alpha g^m - \alpha \beta v^m
$$

From Form II’s update for $v^m$:
$$
v^m = g^{m-1} + \beta v^{m-1}
$$

Recursively substitute until $v^0 = 0$:
$$\boxed{
\therefore v^m = \frac{\theta^m - \theta^{m-1}}{\alpha}}
$$

Substituting this into the update:
$$\boxed{
\theta^{m+1} = \theta^m - \alpha g^m + \beta (\theta^m - \theta^{m-1})}
$$

$\therefore$ $\boxed{\textbf{Form I and Form II give the same update step, proving equivalence by induction.}}$
